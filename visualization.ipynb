```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib.inline
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
plt.plot(x,y,'r')
```

```python
plt.plot(x,y,'b')
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.title("this is my title")
```

```python
plt.scatter(x,y,color='blue',marker="*")
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.title("this is my title")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'b')
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.title("this is my title")
plt.subplot(2,2,4)
plt.scatter(x,y,color='blue',marker="*")
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.title("this is my title")
```

```python
#two graphs in same graphs
fig=plt.figure()
#providing our own axes
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes.plot(x,y,'b',label="bigplot")
axes.legend()
axes.legend(loc="upper right")
axes2.plot(x,y,'red',label="smallerplot")
axes2.scatter(y,x,color="blue",marker="*")
axes2.legend()
axes.set_xlabel("xaxis")
axes.set_ylabel("yaxis")
axes.set_title("this is graph")

axes2.legend(loc="center")
```
